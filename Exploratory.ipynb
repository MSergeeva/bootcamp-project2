```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("data/bank/bank.csv", sep=";")
```

```python
var_mapping = {"yes": 1, "no": 0}
df.replace(var_mapping, inplace=True)
```

```python
df.head()
```

```python
df.describe()
```

```python
categorical = ["job", "education", "marital", "contact", "month"]
```

```python
for col in categorical:
    cat_plot = sns.countplot(df[col], palette="Paired")
    cat_plot.set_xticklabels(df[col].unique(), rotation=45)
    cat_plot.set(xlabel="", ylabel="")
    cat_plot.set_title(col)
    plt.show()
```

```python
interaction_vars = ["poutcome", "default", "loan", "housing", "education", "job"]
```

```python
for col in interaction_vars:
    group_plot = sns.countplot(x="y", hue=col, data=df, palette="Set3")
    group_plot.set(xlabel="", ylabel="")
    group_plot.set_title(col)
    plt.show()
```
